```python
fifa=pd.read_csv('fifa_data.csv')

list(fifa.columns.values)
fifa
```

```python
#dataframe slicing an filter
countries=['Spain','England','Brazil','Portugal']
filt=fifa['Nationality'].isin(countries)
fifa.loc[filt, 'Nationality']
```

```python
#find player with name having Ronaldo
filt=fifa['Name'].str.contains('Ronaldo', na=False)
fifa.loc[filt, 'Name']
```

```python
#find player with score >=90
filt=fifa['Overall']>=90
fifa.loc[filt, ['Name', 'Overall']]
```

```python
bins=[40,50,60,70,80,90,100]
plt.hist(fifa.Overall, bins=bins, color='#abcdef')

plt.xlabel('Skill Level')
plt.ylabel('Number of Players')
plt.title('Distribution of Player Skills in FIFA 2018')
plt.show()

ax=fifa.hist(column='Potential', bins=bins,  color='y')
ax=ax[0]
for x in ax:
    
    x.set_title('Distribution of Player Potential in FIFA 2018')
    x.set_xlabel('Potential Score')
    x.set_ylabel('Number of Players')
    
  
```

```python
nan=pd.isnull(fifa['Height'])
nan_any=fifa.columns[fifa.isnull().any()].tolist()

height=fifa[['Height', 'Overall']]
#drop null rows for height df;
height=height.dropna()
len(height)

def parse_ht(ht):
    ht_=ht.split("'")
    ft_ = float(ht_[0])
    in_ = float(ht_[1])
    return 2.54*((12*ft_)+in_)

height['Height_CM'] =height['Height'].apply(lambda x: parse_ht(x))
height.head()
#height.hist(column='Height_CM', color='b')
```

```python
#check the scatter plot of overall score with height
plt.scatter(height.Height_CM, height.Overall)
from scipy.stats import pearsonr
corr, _ =pearsonr(height.Height_CM, height.Overall)
print('pearson correlation: %.3f' %  corr)
```

```python
left=fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right=fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]
labels=['Left', 'Right']
colors=['#abcdef','#aabbcc']
plt.pie([left, right], labels=labels,colors=colors, autopct='%.2f%%')
#calculate average overall score for preferred foot
fifa.groupby('Preferred Foot').mean()['Overall']
      
```

```python
#draw box chart for left and right foot;
left=fifa.loc[fifa['Preferred Foot']=='Left']['Overall']
right=fifa.loc[fifa['Preferred Foot']=='Right']['Overall']
labels=['Left', 'Right']
boxes=plt.boxplot([left, right], labels=labels, patch_artist=True)
```

```python
fifa.Weight=[int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]
light=fifa.loc[fifa.Weight<125].count()[0]
light_medium=fifa.loc[(fifa.Weight>=125) & (fifa.Weight<150)].count()[0]
medium=fifa.loc[(fifa.Weight>=150) & (fifa.Weight<175)].count()[0]
medium_heavy=fifa.loc[(fifa.Weight>=175) & (fifa.Weight<200)].count()[0]
heavy=fifa.loc[fifa.Weight>=200].count()[0]
light_medium


weight=fifa[['Weight', 'Overall']]
#drop null rows for weight df;
weight=weight.dropna()
len(weight)

corr, _ =pearsonr(weight.Weight, weight.Overall)
print('Pearson correlation: %.3f' % corr)
```

```python
weights=[light, light_medium, medium, medium_heavy, heavy]
labels=['Under 125', '125-150','150-175','175-200','Over 200']
plt.style.use('ggplot')
explode=(0.4,.2,0,0,0.4)
plt.pie(weights, labels=labels, autopct='%.2f%%', explode=explode)
```
