```python
import pandas as pd
import numpy as np

df=pd.read_csv('homeprices2.csv')
df.head()
df.groupby('town').count()
# there are three unquie town value, therefore we need to create three dummies
final=pd.get_dummies(df, drop_first=True)

#create independent variable dataframe;
x=final.drop('price', axis=1)

#create target variable dataframe;
y=final.price
final
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x, y)
model.score(x,y)
#test two examples

model.predict([[3000, 0, 1]])
```
